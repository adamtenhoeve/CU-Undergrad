### Problem A.1

Assume the standard MLR model: for $i = 1,...,n$, $$Y_i = \beta_0 + \beta_1X_{i,1} + ... + \beta_pX_{i,p} + \varepsilon_i,$$ where ${\varepsilon}_i \overset{iid}{\sim}N(0,\sigma^2)$. For simplicity, let's assume that $\sigma^2$ is known. The goal of this question is to show that the least squares estimator is equivalent to the maximum likelihood estimator.

#### (a) Write down the (marginal) pdf of $Y_i$. Call it $f\big(y_i; \, \mu(\boldsymbol\beta), \, \sigma^2 \,\big)$. (HINT: what's the mean and variance of $Y_i$?).


$$f\big( y_i; \mu(\boldsymbol{\beta}), \sigma^2 \big) \sim N(\mu \boldsymbol{\beta}, \sigma^2 I_n) = \frac{1}{\sqrt{2\pi}\sigma}e^{- \frac{1}{2\sigma^2} (Y_i - \beta_0 + \beta_1X_{i,1} + ... + \beta_pX_{i,p})^2}$$

$$
\begin{align*}
f\big( y_i; \mu(\boldsymbol{\beta}), \sigma^2 \big) \sim N(\mu \boldsymbol{\beta}, \sigma^2 I_n) &= \frac{1}{\sqrt{2\pi}\sigma}e^{- \frac{1}{2\sigma^2} (Y_i - \mu(\beta))^2 } \\
&= \frac{1}{\sqrt{2\pi}\sigma}e^{- \frac{1}{2\sigma^2} (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2} \\
\end{align*}
$$

#### (b) Write down the joint pdf of $(Y_1,...,Y_n)$, and simplify your expression using laws for exponentials.  Call it $f_\mathbf{y}\big(\mathbf{y}; \, \mu(\boldsymbol\beta), \, \sigma^2 \,\big)$. (HINT: use the fact that independence allows you to multiply the marginal distributions together.)

$$
\begin{align*}
f_\mathbf{y}\big(\mathbf{y}; \, \mu(\boldsymbol\beta), \, \sigma^2 \,\big) = f_\mathbf{y}(Y_1,...,Y_n; \mu(\beta), \sigma^2) &= \frac{1}{\sqrt{2\pi}\sigma}e^{- \frac{1}{2\sigma^2} (Y_i - \beta_0 + \beta_1X_{i,1} + ... + \beta_pX_{i,p})^2} * ... * \frac{1}{\sqrt{2\pi}\sigma}e^{- \frac{1}{2\sigma^2} (Y_n - \beta_0 + \beta_1X_{n,1} + ... + \beta_pX_{n,p})^2} \\
&= \frac{1}{(2\pi)^{n/2}\sigma^n}e^{- \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2}
\end{align*}
$$

#### (c) The *likelihood function* is the joint pdf of $(Y_1,...,Y_n)$, interpreted as a function of the *parameters* rather than as a function of the data. We might denote the likelihood function as $L(\boldsymbol\beta \, | \mathbf{y} )$ to emphasize that it is a function of the parameters given the data. Write down the log of the likelihood function. Call it $\ell(\boldsymbol\beta \, | \mathbf{y} )$.

Convince yourself that the value of $\boldsymbol\beta$ that maximizes $L$ is the same as the value of $\boldsymbol\beta$ that maximizes $\ell$.

$$
\begin{align*}
L(\boldsymbol{\beta} \mid \boldsymbol{y}) &= \frac{1}{(2\pi)^{n/2}\sigma^n}e^{- \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2} \\
\ell(\boldsymbol{\beta} \mid \boldsymbol{y}) &= log \Big(\frac{1}{(2\pi)^{n/2}\sigma^n}e^{- \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2} \Big) \\
&= \Big(- \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2 \Big) - log\Big( (2\pi)^{n/2}\sigma^n \Big) \\
&= - \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2 + c
\end{align*}
$$

Because the log function is monotonic, the values of $\boldsymbol{\beta}$ that maximize the likelihood function will also be the values that maximize the log of that function.

#### (d) Argue that the maximum likelihood estimator is equivalent to the least squares estimator in this context.

$$
\begin{align*}
    &= max_{\boldsymbol{\beta}}[ - \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2] \\
    &= max_{\boldsymbol{\beta}}[ - \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2] \\
    &= min_{\boldsymbol{\beta}}[ \sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_pX_{i,p})^2] \\
    &= OLS
\end{align*}
$$

The ordinary least squares estimator is the value(s) of $\boldsymbol{\beta}$ that minimizes $\sum_{i=1}^n (Y_i - \beta_0 - \beta_1X_{i,1} - ... - \beta_p X_{i,p})^2$. This is the same summation that will be maximized in the likelihood function above, but negative, once all the constants have been dropped. Because of this, the value of $\boldsymbol{beta}$ that maximizes the likelihood estimator is the same as the least squares estimator that minimizes the ordinary least squares.

#### (c) Now let's try weighted least squares. The code below splits the range of Field into 12 groups of size nine (except for the last goup which has only eight values). Within each group, we compute the variance of Lab as varlab and the mean of Field as meanfield. Add comments to the code to demonstrate what each line is doing.

```r
# Creates index values for each row. 
# Goes in order of the data, as in first row is given index 1, second row is index 2, etc.
i = order(pipeline$Field); 
# Then sorts the data based on Field value, in ascending order. Each row keeps the index given to it in line above.
npipe = pipeline[i,]; 
# Creates 12 groups, each with 9 elements. Each group has a unique number between 1-12.
# Cuts off the last element of the 12th group (108th element). This is because the dataset only has 107 rows.
ff = gl(12,9)[-108]; 
# Split() gets 12 different groups from our sorted Fields data.
# Then get the mean of each of those groups.
# Then unlist turns the dataframe of means into a vector.
meanfield = unlist(lapply(split(npipe$Field,ff),mean))
# Perform the same operation as above but getting the variance of each group.
varlab = unlist(lapply(split(npipe$Lab,ff),var))
```

#### Suppose that the variance in the repsonse is linked to the predictor in the following way: $$ Var(Lab) = a_0Field^{a_2}.$$ Use simple linear regression on (transformations of) varlab and meanfield to estimate $a_0$ and $a_1$. Use these estimates to perform weighted least squares where the weights are the inverse of the variance of Lab. Print a summary of this model and comment on the fit.

To use linear regression, we need a linear relationship between the response and the parameters. Therefor, to solve for $a_0$ and $a_1$, we will need to take the log of the variance.
$$log(Var(Lab)) = a_1log(a_0Field) = a_1log(a_0) + a_1log(Field)$$  
Now we can use our MLR model, were $a_1$ is the slope and $a_0 = e^{\Big(\dfrac{Intercept}{a_1}\Big)}$.

```r
lmod = lm(log(varlab) ~ log(meanfield))
# Get the estimates for a0 and a1 from the model
a1 = as.numeric(lmod$coeff[2])
a0 = exp(as.numeric(lmod$coeff[1]) / a1)
# Create a new weighted model, where the inverse of the variance is the weights
lmod2 = lm(Lab ~ Field, weights = (1/(a0*(Field^a1))), data = pipeline)
summary(lmod2)
par(mfrow = c(2,2)); plot(lmod2)
```

Based on the $R^2$ values of each model, the weighted model (adjusted $R^2 = 0.9188$) is a better fit than the normal model (adjusted $R^2 = 0.8931$). This is because we counteracted the nonconstant variance in Lab data by increasing the weight on larger input Field values than on smaller values. However, can can still note from the plots above that the nonconstant variance in the data was not completely eliminated.
